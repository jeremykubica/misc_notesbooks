```python
import matplotlib.pyplot as plt
import numpy as np
```

# Guess Distances

Plot a visualization of the different guess distances. The gray line out of the Earth indicates the line of site. The gray circles are the different possible object distances from the sun. The black dots are the different possible object locations.

```python
fig =  plt.figure(figsize=(8, 8))
ax = fig.add_subplot()

earth_x = np.cos(np.pi/4)
earth_y = np.sin(np.pi/4)

max_dist = 12
object_angle = 5 * np.pi / 8
object_far_x = max_dist * np.cos(object_angle) + earth_x
object_far_y = max_dist * np.sin(object_angle) + earth_y

# Plot the line from the earth in the direction of the object.
ax.plot([object_far_x, earth_x], [object_far_y, earth_y], color="gray")

# Plot circular orbits coming out from the sun.
angles = np.linspace(0, 2*np.pi, 500)
ax.plot(np.cos(angles), np.sin(angles), color='black')
for radius in range(5, max_dist):
    x = radius * np.cos(angles)
    y = radius * np.sin(angles)
    ax.plot(x, y, color='grey')
    
    # Plot the point on the circle closest to the earth viewing angle
    ang_from_earth = np.arctan2(y - earth_y, x - earth_x)
    ang_dist_from_sight = np.abs(ang_from_earth - object_angle)
    best_fit = np.argmin(ang_dist_from_sight)
    ax.plot(x[best_fit], y[best_fit], color="black", marker=".", markersize=8)

# Plot the sun and the earth    
ax.plot(0, 0, color="yellow", marker=".", markersize=20)
ax.plot(earth_x, earth_y, color="blue", marker=".", markersize=20)
```

# Overhead of the parallax

Plot the Earth and a moving object at two different times, including the line of site between the various objects of interest (Earth, Sun, Asteroid).

```python
fig =  plt.figure(figsize=(8, 8))
ax = fig.add_subplot()

earth_ang1 = 0
earth_ang2 = np.pi / 4
earth1_x = np.cos(earth_ang1)
earth1_y = np.sin(earth_ang1)
earth2_x = np.cos(earth_ang2)
earth2_y = np.sin(earth_ang2)

obj_dist = 4
obj_ang1 = np.pi / 3
obj_ang2 = 7 * np.pi / 16
obj1_x = obj_dist * np.cos(obj_ang1)
obj1_y = obj_dist * np.sin(obj_ang1)
obj2_x = obj_dist * np.cos(obj_ang2)
obj2_y = obj_dist * np.sin(obj_ang2)

# Plot the lines between each pair of objects
ax.plot([0, earth1_x], [0, earth1_y], color='green')
ax.plot([0, obj1_x], [0, obj1_y], color='green')
ax.plot([earth1_x, obj1_x], [earth1_y, obj1_y], color='green')
ax.plot([earth1_x, earth1_x], [earth1_y, earth1_y + 0.5], color='grey')

ax.plot([0, earth2_x], [0, earth2_y], color='red')
ax.plot([0, obj2_x], [0, obj2_y], color='red')
ax.plot([earth2_x, obj2_x], [earth2_y, obj2_y], color='red')
ax.plot([earth2_x, earth2_x], [earth2_y, earth2_y + 0.5], color='grey')


# Plot circular orbits for both objects
angles = np.linspace(0, 2*np.pi, 80)
ax.plot(np.cos(angles), np.sin(angles), color='gray')
ax.plot(obj_dist * np.cos(angles), obj_dist * np.sin(angles), color='gray')



# Plot the sun and the earth    
ax.plot(0, 0, color="yellow", marker=".", markersize=20)
ax.plot(earth1_x, earth1_y, color="blue", marker=".", markersize=10)
ax.plot(earth2_x, earth2_y, color="blue", marker=".", markersize=10)
ax.plot(obj1_x, obj1_y, color="black", marker=".", markersize=10)
ax.plot(obj2_x, obj2_y, color="black", marker=".", markersize=10)
```
